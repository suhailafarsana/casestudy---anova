# ● Check whether all the zones generate the same amount of sales.

```python
#H0-all the zones generate the same amount of sales
#H1-all the zones generate the different amount of sales
fvalue,pvalue=stats.f_oneway(df['Zone - A'],df['Zone - B'],df['Zone - C'],df['Zone - D'])
print('f value:',fvalue, 'p value:',pvalue)

if pvalue>0.05:
    print("accept H0")
else:
    print("reject H0")
```

that means all the zones generate the different amount of sales
